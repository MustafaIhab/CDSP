```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
# Making connection
database = 'books.db'
connection = sql.connect(database)
```

```python
#Write a sql query to get books that has rate >=3 and has 'Mr' in its name.
query = 'SELECT * FROM books WHERE rate >= 3 AND title like "%Mr%"'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to get books that has rate >=3 or price > 20.
query = 'SELECT * FROM books WHERE rate >= 3 or price > 20'
pd.read_sql_query(query, connection).head()
```

```python
#Write a sql query to get books that has rate not 3.
query = 'SELECT * FROM books WHERE rate != 3'
pd.read_sql_query(query, connection).head()
```

```python
#Write a sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.
query = 'SELECT * FROM books WHERE (price BETWEEN 10 AND 40) AND (rate = 3 or rate = 4 or rate = 1)'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to get the top 5 most expensive books (order books desc with price column and then limit the result to the first 5)
query = 'SELECT * FROM books ORDER BY price DESC LIMIT 5'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to get the 3rd 10 books in the books table order first by rate desc and then by price asc.
query = 'SELECT * FROM books ORDER BY rate DESC, price ASC LIMIT 10 OFFSET 10*2'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to add a new category in categories table and add 5 books to this category in books table with title, rate and price.
query = 'INSERT INTO categories(ID, Name) VALUES(51, "Manga")'
connection.execute(query)
```

```python
#Write a sql query to update book's rate to 3 that thier price < 20£.
query = 'UPDATE books SET rate = 3 WHERE price < 20'
connection.execute(query)

#TEST
pd.read_sql_query('SELECT * FROM books WHERE price < 20', connection)
```

```python
#Write a sql query to delete all books that have price > 50£ and has rate <= 2.
query = "DELETE FROM books WHERE price > 50 and rate <= 2"
connection.execute(query)

#TEST
pd.read_sql_query('SELECT * FROM books WHERE price > 50 and rate <=2', connection)
```

```python
#Write a sql query to count the number of books that have 'Secret' in thier names and price between 10£ and 25£.
query = 'SELECT COUNT(title) as "number of books" FROM books WHERE (title like "%Secret%") AND (price BETWEEN 10 AND 25)'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to get the minimum & maximum price for all the books that have rate 5.
query = 'SELECT MAX(price) as "Max Price", MIN(price) as "Min price" FROM books WHERE rate = 5'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to calculate the avg price for all the books that have rate 5.
query = 'SELECT AVG(price) as "average price" FROM books WHERE rate = 5'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to sum all book's price that have rate 2 and price between 10 and 40
query = 'SELECT SUM(price) as "SUM price" FROM books WHERE (rate = 2) AND (price BETWEEN 10 AND 40)'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.
query = 'SELECT books.title, books.rate, books.price, categories.name as category FROM books INNER JOIN categories ON books.category_id=categories.id'
pd.read_sql_query(query, connection).head()
```

```python
#Write a sql query to calculate how many books each rate has and have price between 20£ and 30£
query = 'SELECT rate as RATE, count(title) as "COUNT" FROM books WHERE price BETWEEN 20 AND 30 GROUP BY rate ORDER BY rate'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to calculate how many books each category has having count > 10.
query = 'SELECT category_id, count(title) as count FROM books GROUP BY category_id HAVING count > 10'
pd.read_sql_query(query, connection)
```

```python
#Write a sql query to get all books with category_name='Music' using subquery.
query = 'SELECT * FROM books b, categories c WHERE b.category_id = c.id AND c.name = "Music"'
pd.read_sql_query(query, connection)
```
